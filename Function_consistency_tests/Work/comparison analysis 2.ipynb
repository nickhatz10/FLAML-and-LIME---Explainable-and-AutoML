```python
from plotnine import ggplot, aes, geom_line, geom_smooth, geom_point

(
    ggplot(record_df)  # What data to use
    + aes(x="row_num", y="f1_score")  # What variable to use
    + geom_point()
    + geom_smooth(colour = "red")  # Geometric object to use for drawing
)
```
